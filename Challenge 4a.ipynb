### Challenge 4

First, create a function that will check if a number (a) is a factor of another number (b).  The function should return a value of True if a is a factor of b.

Then, create another function that will calculate all of the factors of a given number.  You may use your function from part 1 above in this function, or use entirely new logic. This function should return a list of all the factors of the number.

Finally, use your function(s) in a loop that will print out each number from 1 to 100 and all of the factors of those numbers.  The printout should look something like this:

`1 - [1]`

`2 - [1, 2]`

`3 - [1, 3]`

...

`100 - [1, 2, 4, 5, 10, 20, 25, 50, 100]`

```python
# Write code here

def factCheck (a, b):
    if b%a == 0:
        return True
    return False

def allFacts (num):
    factList = []
    for n in range (1, num+1):
        if factCheck(n, num):
            factList.append(n)
    return factList

print(factCheck(2,5))
for i in range (1, 101):
    print(f"{i} - {allFacts(i)}")
```
