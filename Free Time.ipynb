```python
java
package frc.robot.commands;

import edu.wpi.first.wpilibj2.command.SequentialCommandGroup;
import edu.wpi.first.wpilibj2.command.WaitCommand;
import frc.robot.subsystems.CANDrivetrainSubsystem;
import frc.robot.subsystems.CANLauncherSubsystem;

import static frc.robot.Constants.LauncherConstants.kLaunchFeederSpeed;
import static frc.robot.Constants.LauncherConstants.kLauncherSpeed;

public class AutoLaunch extends SequentialCommandGroup {

    public AutoLaunch(CANLauncherSubsystem canLauncherSubsystem, CANDrivetrainSubsystem canDrivetrainSubsystem, double driveTime, double prepTime, double finishDelay) {
        addCommands(
            new PrepareLaunch(canLauncherSubsystem),
            new WaitCommand(prepTime),
            new LaunchNote(canLauncherSubsystem).withTimeout(finishDelay),
            new AutoDrive(canDrivetrainSubsystem, driveTime)
        );
    }
}
```
